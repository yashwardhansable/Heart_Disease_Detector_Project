```python
df = pd.read_csv('heart_2020_cleaned.csv')
```

```python
smoking_dummies = pd.get_dummies(df['Smoking'],drop_first=True)
```

```python
df = pd.concat([df.drop(['Smoking'], axis = 1),smoking_dummies],axis = 1)
```

```python
alc_dummies = pd.get_dummies(df['AlcoholDrinking'],drop_first=True)
```

```python
stroke_dummies = pd.get_dummies(df['Stroke'],drop_first=True)
```

```python
walk_dummies = pd.get_dummies(df['DiffWalking'],drop_first=True)
```

```python
dia_dummies = pd.get_dummies(df['Diabetic'],drop_first=True)
```

```python
sex_dummies = pd.get_dummies(df['Sex'],drop_first=True)
```

```python
physical_dummies = pd.get_dummies(df['PhysicalActivity'],drop_first=True)
```

```python
asthama_dummies = pd.get_dummies(df['Asthma'],drop_first=True)
```

```python
kid_dummies = pd.get_dummies(df['KidneyDisease'],drop_first = True)
```

```python
sk_dummies = pd.get_dummies(df['SkinCancer'],drop_first=True)
```

```python
Age_dummies = pd.get_dummies(df['AgeCategory'],drop_first=True)
```

```python
Race_dummies = pd.get_dummies(df['Race'],drop_first=True)
```

```python
Gen_dummies = pd.get_dummies(df['GenHealth'],drop_first=True)
```

```python
df = df.drop(['SkinCancer'], axis =1)
df = df.drop(['KidneyDisease'],axis = 1)
df = df.drop(['Asthma'],axis = 1)
df = df.drop(['PhysicalActivity'],axis = 1)
df = df.drop(['Sex'],axis = 1)
df = df.drop(['Diabetic'],axis = 1)
df = df.drop(['DiffWalking'],axis = 1)
df = df.drop(['Stroke'],axis = 1)
df = df.drop(['AlcoholDrinking'],axis = 1)
df = df.drop(['AgeCategory'],axis =1)
df = df.drop(['Race'],axis =1)
df = df.drop(['GenHealth'], axis =1)
```

```python
df = pd.concat([df,Gen_dummies],axis = 1)
df = pd.concat([df,Race_dummies],axis = 1)
df = pd.concat([df,Age_dummies],axis = 1)
df = pd.concat([df,alc_dummies,stroke_dummies,walk_dummies,sex_dummies,dia_dummies,physical_dummies,asthama_dummies,kid_dummies],axis = 1)
```

```python
df = pd.concat([df,smoking_dummies],axis = 1)
```

```python
df
```

```python
Heart_dummies = pd.get_dummies(df['HeartDisease'],drop_first=True)
```

```python
df['TARGET'] = Heart_dummies
```

```python
df = df.drop('HeartDisease', axis =1)
```

```python
df
```

```python
minority_class_len = len(df[df['TARGET']==1])
majority_class_indices = df[df['TARGET']== 0].index
random_majority_indices = np.random.choice(majority_class_indices,
                                          minority_class_len,
                                          replace = False)
minority_class_indices = df[df['TARGET']==1].index
under_sample_indices = np.concatenate([minority_class_indices,random_majority_indices])
under_sample = df.loc[under_sample_indices]
```

```python
X_under = under_sample.drop(['TARGET'], axis = 1)
```

```python
y_under = under_sample['TARGET']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_under, y_under, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# tensor flow model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.constraints import max_norm
```

```python
model = Sequential()

# https://stats.stackexchange.com/questions/181/how-to-choose-the-number-of-hidden-layers-and-nodes-in-a-feedforward-neural-netw


# input layer
model.add(Dense(37,  activation='ReLU'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(18, activation='sigmoid'))
model.add(Dropout(0.2))

# hidden layer
model.add(Dense(9, activation='ReLU'))
model.add(Dropout(0.2))

# output layer
model.add(Dense(units=1,activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=35,
          batch_size=256,
          validation_data=(X_test, y_test), 
          )
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses[['loss','val_loss']].plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred = model.predict(X_test)
y_pred = np.round(y_pred).astype(int)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
